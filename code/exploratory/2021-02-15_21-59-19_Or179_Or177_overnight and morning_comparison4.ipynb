````python
# Set up widgets
wav_selector = pnw.Select(options=list(range(len(wav_df))), name="Select song file")
offset_selector = pnw.Select(options=np.linspace(-10,10,21).tolist(), name="Select offset")
window_radius_selector = pnw.Select(options=[0,10,20,30,40,60], name="Select window radius")
spect_chan_selector = pnw.Select(options=list(range(16)), name="Spectrogram channel")
spect_freq_lo = pnw.Select(options=np.linspace(0,400,41).tolist(), name="Low frequency for spectrogram (Hz)")
spect_freq_hi = pnw.Select(options=np.linspace(400,0,41).tolist(), name="Hi frequency for spectrogram (Hz)")
log_nfft_selector = pnw.Select(options=np.linspace(8,16,9).tolist(), name="magnitude of nfft (starts at 256)")

@pn.depends(
    wav_selector=wav_selector.param.value,
    offset=offset_selector.param.value,
    window_radius=window_radius_selector.param.value,
    spect_chan=spect_chan_selector.param.value,
    spect_freq_lo=spect_freq_lo.param.value,
    spect_freq_hi=spect_freq_hi.param.value,
    log_nfft=log_nfft_selector.param.value
)
def create_figure(
    wav_selector,
    offset,
    spect_chan,
    spect_freq_lo,
    spect_freq_hi,
    log_nfft,
):
    # Each column in each row to a tuple that we unpack
    wav_file_path, wav_file_name, tr0, tr1 = wav_df.loc[wav_selector,:]
    
    print(wav_file_name)
    
    # Set up figure
    fig,axes = plt.subplots(5,1, figsize=(12,8))
    
    # Apply offset and apply window radius
    tr0 += offset-window_radius
    tr1 += offset+window_radius
    
    '''Plot Hi fi mic'''
    # Get wav file numpy recording object
    wav_recording = get_wav_recording(wav_file_path)
    sw.plot_spectrogram(wav_recording, channel=0, freqrange=[300,14000],ax=axes[0])
    
    '''Plot Lo fi mic ''' 
    if 'Or179' in wav_file_name:
        mic_recording = Or179_recording
        mic_channel = Or179_recording.get_channel_ids()[-1]
    elif 'Or177' in wav_file_name:
        mic_recording = Or177_recording
        mic_channel = Or177_recording.get_channel_ids()[-1]
    
    sw.plot_spectrogram(
        mic_recording,
        mic_channel,
        trange=[tr0, tr1],
        freqrange=[300,12000],
        ax=axes[1]
    )
    )
    
    '''Plot timeseries'''
    chan_ids = np.array([mic_recording.get_channel_ids()]).flatten()
    sw.plot_timeseries(
        mic_recording,
        channel_ids=chan_ids[1:4],
        trange=[tr0, tr1],
        ax=axes[2]
    )
    # Clean lines
    for line in plt.gca().lines:
        line.set_linewidth(0.5)    
    
    '''Plot log power timeseries'''
    mic_recording.get_traces(
        channel_ids=chan_ids, 
        start_frame=tr0*mic_recording.get_sampling_frequency(), 
        end_frame=tr1*mic_recording.get_sampling_frequency()
    ).flatten()
    
    '''Plot LFP spectrogram'''
    sw.plot_spectrogram(
        mic_recording,
        channel=chan_ids[spect_chan],
        freqrange=[spect_freq_lo,spect_freq_hi],
        trange=[tr0, tr1],
        ax=axes[3],
        nfft=int(2**log_nfft)
    )
    
    # Remove axes
    [ax.axis('off') for ax in axes]
    
    return fig
````
